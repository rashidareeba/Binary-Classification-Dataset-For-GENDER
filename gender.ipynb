```python
df['GENDER'] = df['GENDER'].map({1: 0, 2: 1})
```

```python
df.head(108)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['HEIGHT'], df['HAND_LENGTH'], c=df['GENDER'], cmap='viridis', alpha=0.7)
```

```python
X = df[['HEIGHT', 'HAND_LENGTH', 'FOOT_LENGTH']]
y = df['GENDER']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)    
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['HEIGHT'], df['HAND_LENGTH'], c=df['GENDER'], cmap='viridis', alpha=0.7)
plt.xlabel('Height')
plt.ylabel('Hand Length')
plt.title('Gender classification based on Height and Hand Length')
plt.colorbar(label='Gender')  
plt.show()
```

```python
import sklearn.metrics as f1
from sklearn.metrics import f1_score
f1_score_value = f1_score(y_test, y_pred)
print(f"F1 Score: {f1_score_value:.4f}")
plt.bar(['F1 Score'], [f1_score_value], color='blue')
plt.ylabel('Score')
plt.title('F1 Score of the Model')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()  
print(classification_report(y_test, y_pred))
```
